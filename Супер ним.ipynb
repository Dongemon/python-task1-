```python
import random


#Функция создает списки со случайным количеством фишек
def field():
    field = []
    for i in range(8):
        field.append(random.sample(['O', ' '], counts=[8, 40], k = 8))
    return field


#Функция рисует доску
def print_field(field):
    print('\n  | a | b | c | d | e | f | g | h |')
    for i in range(8):
        print('-----------------------------------')
        print(f'{i + 1} |', ' | '.join(field[i]), '|')
    
        
#Функция проверяет, есть ли фишки на доске
def check_win(field):
    for i in field:
        if i.count('O') != 0:
            return False
    return True


#Удаляет фишки с вертикали
def delete_vertical(letter):
    
    index = letters.index(letter)
    
    for i in range(8):
        field[i][index] = ' '
        

#Удаляет фишки с горизонтали
def delete_gorizontal(number):
    
    for i in range(8):
        field[int(number) - 1][i] = ' '
        
        
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
numbers = ['1', '2', '3', '4', '5', '6', '7', '8']

field = field()

while True:
    
    print_field(field)
    
    while True:
        player1 = input('Игрок 1\nВыберите вертикаль или горизонталь, с которой хотите убрать фишки: ')
        if player1 in letters or player1 in numbers:
            break
        else:
            print('Я не могу это сделать(')
    
    
    if player1 in letters:
        delete_vertical(player1)
    else:
        delete_gorizontal(player1)
    
    if check_win(field):
        print('Игра окончена!\nИгрок 1 победил!')
        break
        
    print_field(field)
            
    while True:
        player2 = input('Игрок 2\nВыберите вертикаль или горизонталь, с которой хотите убрать фишки: ')
        if player2 in letters or player2 in numbers:
            break
        else:
            print('Я не могу это сделать(')
    
    
    if player2 in letters:
        delete_vertical(player2)
    else:
        delete_gorizontal(player2)
    
    if check_win(field):
        print('Игра окончена!\nИгрок 2 победил!')
        break
           
    
```
