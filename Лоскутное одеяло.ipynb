```python
#Создает массив для поля
def field():
    field = []
    for i in range(4):
        field.append([' ' for i in range(5)])
    return field


def print_field(score1, score2, field):
    print(f'---------------------------\
\n| Игрок1 - {score1} | Игрок2 - {score2} |\
\n---------------------------')
    print('   | a | b | c | d | e ')
    for i in range(4):
        print('-----------------------')
        print(f' {i + 1} |', ' | '.join(field[i]))

```

```python
#Функция, подсчитывающая количество очков после хода игрока
def check_score(field, move, key):
    score = 0
    
    if letters.index(move[0]) == 4:
        y = [3, 4]
    elif letters.index(move[0]) == 0:
        y = [0, 1]
    else:
        y = [letters.index(move[0]) - 1, letters.index(move[0]), letters.index(move[0]) + 1]
    
    if numbers.index(move[1]) == 3:
        x = [2, 3]
    elif numbers.index(move[1]) == 0:
        x = [0, 1]
    else:
        x = [numbers.index(move[1]) - 1,numbers.index(move[1]), numbers.index(move[1]) + 1]
    
    for i in x:
        for j in y:
            if field[i][j] == key:
                score += 1
   
    return score - 1
```

```python
#Изменяет массив-поле, фиксируя ход игрока
def do_move(feild, move, key):
    field[numbers.index(move[1])][letters.index(move[0])] = key
```

```python
field = field()

numbers = ['1', '2', '3', '4']
letters = ['a', 'b', 'c', 'd', 'e']
score1 = 0
score2 = 0
moves = 0
played_moves = []

while True:
    
    print_field(score1, score2, field)
    while True:
        player1 = input('Игрок 1\nВыберите клетку(сначала буква, потом цифра): ')
        if player1[0] in letters and player1[1] in numbers and len(player1) == 2 and player1 not in played_moves:
            break
        else:
            print('Ход формата буква-цифра(Например:a1, a2) и нельзя ходить повторно в ту же клетку.')
    
    played_moves.append(player1)
    
    do_move(field, player1, 'X')
    
    score1 += check_score(field, player1, 'X')
    
    print_field(score1, score2, field)
    
    while True:
        player2 = input('Игрок 2\nВыберите клетку(сначала буква, потом цифра): ')
        if player2[0] in letters and player2[1] in numbers and len(player2) == 2 and player2 not in played_moves:
            break
        else:
            print('Ход формата буква-цифра(Например:a1, a2) и нельзя ходить повторно в ту же клетку.')
    
    do_move(field, player2, 'O')
    
    score2 += check_score(field, player2, 'O')
    
    moves += 1
    if moves == 10:
        break
        
if score1 > score2:
    print(f'Игра завершена со счетом {score1} - {score2}.\nПобедил игрок 2')
elif score2 > score1:
    print(f'Игра завершена со счетом {score1} - {score2}.\nПобедил игрок 1')
else:
    print(f'Игра завершена со счетом {score1} - {score2}.\nНичья!')
```
