```python
import random


def field():
    field = []
    for i in range(3):
        field.append([str(random.randint(1,9)) for i in range(3)])
    return field


def print_field(field, sum1, sum2):
    print(f'-----------------------------\n| Игрок 1 - {sum1} | Игрок 2 - {sum2} |\n-----------------------------')
    print('  |   | a | b | c |')
    for i in range(3):
        print('-------------------')
        print(f'{i + 1} |   |', ' | '.join(field[i]), '|')
    print()
```

```python
def do_move(field, move, key):
    field[move[0]][move[1]] = key
```

```python
def check_moves(field, move, key):
    possible_moves = []
    
    if key == 'O':
        
        if field[move[0]][(move[1] + 1) % 3] not in ['O', 'X']:
            possible_moves.append([move[0], (move[1] + 1) % 3])
        if field[move[0]][(move[1] + 2) % 3] not in ['O', 'X']:
            possible_moves.append([move[0], (move[1] + 2) % 3])
            
    elif key == 'X':
        
        if field[(move[0] + 1) % 3][move[1]] not in ['O', 'X']:
            possible_moves.append([(move[0] + 1) % 3, move[1]])
        if field[(move[0] + 2) % 3][move[1]] not in ['O', 'X']:
            possible_moves.append([(move[0] + 2) % 3, move[1]])
            
    return possible_moves
```

```python
sum1 = 0
sum2 = 0

letters = ['a', 'b', 'c']
numbers = ['1', '2', '3']
```

```python
possible_moves1 = [[0, 0], [0, 1],[0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2]]
possible_moves2 = [[0, 0], [0, 1],[0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2]]
played_moves = []
field = field()
```

```python
while True:
    
    print_field(field, sum1, sum2)
    
    #Ход первого игрока
    while True:

        if len(possible_moves1) == 0:
            possible_moves2 = check_moves(field, move, 'X')
            print('Игрок 1\nУ вас нет ходов(вся строка занята).')
            break
        player1 = input('Игрок 1\nВыберите клетку(сначала буква, потом цифра): ')
        if player1[0] in letters and player1[1] in numbers and len(player1) == 2 and player1 not in played_moves:
            move = [numbers.index(player1[1]), letters.index(player1[0])]
            
            if move in possible_moves1:
                
                sum1 += int(field[move[0]][move[1]])
                played_moves.append(player1)
                do_move(field, move, 'X')
                possible_moves2 = check_moves(field, move, 'X')
                break
            else:
                print('Вы можете только в клетки строки последнего хода.')
        else:
            print('Ход формата буква-цифра(Например:a1, a2) и нельзя ходить повторно в ту же клетку.')
    
    print_field(field, sum1, sum2)
    
    #Ход второго игрока
    while True:
        
        if len(possible_moves2) == 0:
            possible_moves1 = check_moves(field, move, 'O')
            print('Игрок 2\nУ вас нет ходов(весь столбец занят).')
            break
        player2 = input('Игрок 2\nВыберите клетку(сначала буква, потом цифра): ')
        if player2[0] in letters and player2[1] in numbers and len(player2) == 2 and player2 not in played_moves:
            move = [numbers.index(player2[1]), letters.index(player2[0])]
            
            if move in possible_moves2:
                
                sum2 += int(field[move[0]][move[1]])
                played_moves.append(player2)
                do_move(field, move, 'O')
                possible_moves1 = check_moves(field, move, 'O')
                break
            else:
                print('Вы можете только в клетки столбца последнего хода.')
        else:
            print('Ход формата буква-цифра(Например:a1, a2) и нельзя ходить повторно в ту же клетку.')
        
    if len(possible_moves1) == 0 and len(possible_moves2) == 0:
        if sum1 > sum2:
            print('Игра окончена.\nИгрок 1 победил!')
            break
        elif sum2 > sum1:
            print('Игра окончена.\nИгрок 2 победил!')
            break
        else:
            print('Игра окончена.\nНичья!')
            break
```
